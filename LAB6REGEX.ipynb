### El punto

•Se interpreta como "Cualquier carácter" sin incluir el salto de línea.

• Ejemplo:
• Si se dice al motor de Regex que busque "g.t" en la cadena, "el gato de piedra gótica puerta de getisboro goot"
El motor encontrará "gat", "gót", "get", pero no encontrará "goot" porque el punto equivale a un solo carácter.

### La barra inversa \

•Se utiliza para "marcar" el siguiente carácter de la expresión de búsqueda de forma que este adquiera un significado especial o deje de tenerlo
• Ejemplo:
Si utilizamos "\.", el motor encontrará un punto y no como vimos anteriormente que el punto representa cualquier carácter.

 La barra inversa "\":
 * \d — Representa un dígito del 0 al 9.
 * \w — Representa cualquier carácter alfanumérico.
 * \s — Representa un espacio en blanco.
 * \D — Representa cualquier carácter que no sea un dígito del 0 al 9.
 * \W — Representa cualquier carácter no alfanumérico.
 * \S — Representa cualquier carácter que no sea un espacio en blanco.
 * \A — Representa el inicio de la cadena. No un carácter sino una posición.
 * \Z — Representa el final de la cadena. No un carácter sino una posición.
 * \b - Representa los "limites" de la expresion regular.

### Los corchetes "[ ]"
•La función de los corchetes es representar "clases de caracteres", osea, agrupar caracteres en grupos o clases, tambien podemos utilizar el "-" para especificar rangos.

• Ejemplo:

* "[a-z]" el motor encuentra un carácter de la "a" a la "z" 
* "[0-9]" el motor encuentra un carácter entre "0" y "9"

Nota: Los caracteres especiales pierden su significado y se convierten en literales al estar adentro de corchetes.


 ### La barra "|"
• La barra sirve para indicar una entre varias opciones.
• Ejemplo:

La expresión regular "a|e" encontrará "a" o "e" dentro del texto.

La expresión regular "este|oeste|norte|sur" permitirá encontrar cualquiera de los nombre de los puntos cardinales.

### El dólar "$" y El acento circunflejo "^"

* El dólar representa el final de una cadena de caracteres o el final de una linea.

* El acento circunflejo tiene dos funcionalidades,
  * 1. Representa el inicio de una linea (así como el dólar representa el
final) .
  * "^[a-z]" encuentra parrafos que empiecen con letra minuscula.
  * 2. Si se utiliza adentro de corchetes, representa la negación de todo lo que se encuentre en ese grupo.
  * "[^\w]" encuentra cualquier carácter que NO se encuentre en el grupo
indicado.

### Los paréntesis "( )"
* Similar a los corchetes, los paréntesis sirven para agrupar caracteres, aquí los caracteres especiales conservan su función.
* Ejemplo: "al (este|oeste|norte|sur) de" permite buscar textos que den indicaciones por medio de puntos cardinales.


### Signo de interrogación "?"
* Especifica que una parte de la búsqueda es opcional.
* Ejemplo: "ob?scuridad", encontrará "obscuridad" y "oscuridad".
* Utilizado en con paréntesis especifica que un conjunto mayor de caracteres es opcional.


### El asterisco "*"
* El asterisco es utilizado para encontrar algo que se encuentra repetido 0 o más veces.
* Ejemplo:
* "[a-zA-Z]\d*" encontrará "H","H0","H01","h0100010"

 ### bEl signo suma "+"
* El signo suma es utilizado para encontrar algo que se encuentra repetido 1 o más veces.
* Ejemplo:
* "[a-zA-Z]\d+" encontrará "H0","H01","h0100010" PERO no "H"

#### 3.	Genere una expresión regular para validar contraseñas de correo. Una contraseña de correo debe contener por lo menos 8 caracteres, una letra mayúscula y un carácter especial.

```python
string_3 = "Marroquin$ lapicero"
pattern_dot = "(?=.{8,})(?=.*[A-Z])(?=.*[@#$%^*()+-])[\w@#$%^*()+-]{8,}"
```

#### 6.	Cree una expresión regular para obtener los números telefónicos de Guatemala. Estos pueden contener al inicio +502 o 502, pueden estar separados por un espacio en blanco o un guión o juntos. Notar que los números telefónicos pueden empezar únicamente con 4,5,6 o 2.
#### a.	+50254821151, 4210-7640, 52018150, 2434 6854, 11234569, 50211234578


```python
string_6 = "+50254821151, 4210-7640, 52018150, 2434 6854, 11234569, 50211234578"
pattern_dot = "(?:\+502|502)?[2-6]\d{3}[-\s]?\d{4}"
```

```python
re.findall(pattern_dot,string_6)
```
